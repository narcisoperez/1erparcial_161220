```python
import numpy
import matplotlib.pyplot as plt

from pyearth import Earth

# Create some fake data
numpy.random.seed(2)
m = 10000
n = 10
X = 80 * numpy.random.uniform(size=(m, n)) - 40
y = 100 * \
    (numpy.sin((X[:, 6])) - 4.0) + \
    10 * numpy.random.normal(size=m)

# Fit an Earth model
model = Earth(max_degree=3, minspan_alpha=.5, verbose=True)
model.fit(X, y)

# Print the model
print(model.trace())
print(model.summary())

# Plot the model
y_hat = model.predict(X)
plt.figure(figsize=(16,10))
plt.plot(X[:, 6], y, 'r.')
plt.plot(X[:, 6], y_hat, 'b.')
plt.show()
```

### Como se graficaría una función con discontinuidad

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

V = 1000.0 # L
nu = 6.0  # L/min
    
def Cs_in(t):
    'inlet concentration'
    if t < 0:
        Cs = 0.0 # g/L
    elif (t > 0) and (t <= 10):
        Cs = 4.0
    else:
        Cs = 2.0
    return Cs

def mass_balance(Ms, t):
    '$\frac{dM_S}{dt} = \nu C_{S,in}(t) - \nu M_S/V$'
    dMsdt = nu * Cs_in(t) - nu * Ms / V
    return dMsdt

tspan = np.linspace(-10.0, 15.0, 100)

M0 = 300.0 # gm salt
Ms = odeint(mass_balance, M0, tspan)
plt.figure(figsize=(15,10))
plt.plot(tspan, Ms/V, 'b.-')
plt.xlabel('Time (min)')
plt.ylabel('Salt concentration (g/L)')
```

$f(0) = 0 \hspace{0.5cm} lim_{x \to 0^-} 0 = 0 \hspace{0.5cm} lim_{x \to 0^+} x = 0$

## b) Toda función derivable en un punto es continua en ese punto.

## $\text{Verdadero.}$

$\text{Demostración:}$

$\text{Por definición sabemos que: }$

$lim_{h \to 0} \frac{f(a+h) - f(a)}{h} = f'(a)$, entonces $lim_{x \to a} f(x) = f(a) $

$lim_{x \to a} f(x) = f(a) \Rightarrow lim_{h \to 0} f(a + h) = f(a)$

$\text{Por álgebra: }$

$lim_{h \to 0}f(a+h)-f(a) = lim_{h \to 0} h \cdot \frac{f(a + h) - f(a)}{h}$

$\text{Aplicando defnición de límite: }$

$lim_{h \to 0}f(a+h)-f(a) = 0 \cdot f'(a) \Rightarrow lim_{h \to 0}f(a+h)-f(a) = 0$
